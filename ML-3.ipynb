```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ran = np.random.RandomState(7)
x = 10*ran.randn(50)
y = 2*x+3+ran.randn(50)
x, y
```

```python
y[10] = 150
y[5] = 100
y[15] = 105

plt.figure(1)
plt.boxplot(y)
plt.figure(2)
plt.scatter(x, y)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)

from sklearn.linear_model import LinearRegression
from IPython.display import Math, display, Latex

model = LinearRegression()
model.fit(x, y)
display(Math("\\beta_o = "+(str(model.intercept_[0]))))
display(Math("\\beta_1 = "+str(model.coef_[0][0])))

y_pred = model.predict(x)

plt.scatter(x, y, color='g')
plt.plot(x, y_pred, color='r')
plt.show()
```
