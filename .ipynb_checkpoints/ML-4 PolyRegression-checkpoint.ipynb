```python
from IPython.display import Math, display
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
```

```python
rand = np.random.RandomState(7)
x = 10*rand.randn(50)
y = np.sin(x)+0.1*rand.randn(50)

import matplotlib.pyplot as plt

plt.title("NON-LINEAR CURVE")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y)
plt.show()
```

```python
#Using Lear
from sklearn.linear_model import LinearRegression

x = x.reshape(-1,1)
y = y.reshape(-1,1)
model = LinearRegression()
model.fit(x, y)

y_predict = model.predict(x)

plt.plot(x, y_predict, color='r', label='Predicated')
plt.scatter(x, y, color='g', label='Actual')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=10, include_bias=False)
#Adjust the degree using hit and trial method to find the best fix
x_poly = poly.fit_transform(x)
#Transforming to polynomial i.e finds relation b/w x & y
poly.fit(x_poly, y)

model = LinearRegression()
model.fit(x_poly, y)

y_pred = model.predict(poly.fit_transform(x))
plt.scatter(x, y_pred, color='r')
plt.scatter(x, y, color='g', label='Actual')

plt.yticks(np.arange(-2,3))

plt.show()
```

```python
df = pd.read_csv("Position_Salary.csv")

X = df.iloc[:,1:2]
Y = df.iloc[:,-1]

x = X.to_numpy()
y = Y.to_numpy()

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)

y_test = np.array(y_test, dtype=float)

y_test, y_predict
```

```python
#Mean absolute Error 1/nsum\y-y'\
#Root mean Square Error sqrt(1/nsum(y-y')^2)
#Root mean squared logarithmic error sqrt(1/nsum)
#R-Square Error = 1 - MSE(model)/MSE(Baseline)
    #MSE(MODEL) = sum\y-y'\
    #MSE(Baseline) = sum(y-(bar)-y')^2
    
#Adjusted-R Square = 1-(1-R^2)[(n-1)/(n-(k+1))]
# n - total number of feature
# k - total number of sample

display(Math("\\text{Mean Absolute Error}=\\frac{1}{n}\\sum|y-y'|"))

display(Math("\\text{Root mean Square Error sqrt}=\\sqrt{\\frac{1}{n}\\sum|y-y'|^2}"))

display(Math("\\text{Root mean squared logarithmic error}=\\sqrt{\\frac{1}{n}\\sum{(\\log(y+1)-\\log(y'+1))^2}}"))

display(Math("\\text{R-Square Error} = 1-\\frac{\\text{MSE(Model)}}{MSE(Baseline)}"))

display(Math("\\text{MSE(Model)} = \\sum{|y-y'|}"))

display(Math("\\text{MSE(Baseline)} = \\sum(y^{-} - y')^2"))

display(Math("\\text{Adjusted-R Square}=1-(1-R^2)\\begin{bmatrix}\\frac{n-1}{n-(k+1)}\\end{bmatrix}"))
#Matrix is used to have larger bracket.

display(Math("\\text{n} ->\\text{Total number of feature}"))

display(Math("\\text{k} ->\\text{Total number of sample}"))
```

```python
import numpy as np
import math

x = np.arange(-10,11)
y = 1/(1+np.power(math.e, -x))

plt.title("Sigmod Function")
plt.xlabel("Value of x")
plt.ylabel("Sigmoid Value")
plt.xticks(x)
plt.yticks(np.arange(0.1,1.1,0.1))
plt.grid(True)
plt.plot(x, y)
plt.show()
```
